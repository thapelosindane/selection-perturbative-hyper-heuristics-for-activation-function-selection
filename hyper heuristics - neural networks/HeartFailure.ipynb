```
# hyper heuristics

'''
WINE LOG
15 NODES ON FIRST LAYER
8 HEURISTICS
40 EPOCHS

'''
import random
heuristics = ["h1","h2","h3","h4"]
def get_initial_state() :
  """generate the initial state of the system"""
  random.seed(31)
  global heuristics
  #heuristics = ["h1","h2","h3","h4"]
  heuristic_strings = []
  for i in range(8):
      strng = random.choice(heuristics)
      heuristic_strings.append(strng)
  return heuristic_strings    



# neighbors
def get_neighbors(heuristic_strings):
     global heuristics
     neighbors = []
     def low_heur_to_change():
               h_chnged = random.choice(heuristic_strings)
               index = heuristic_strings.index(h_chnged)
               return index

     for i in range(len(heuristic_strings)):          
            for id in range(4):
                  
                  index = low_heur_to_change()
                  h_changed = heuristic_strings[index]
                  list_to_add = []
                  for heu in heuristics:
                      if heu !=h_changed:
                        list_to_add.append(heu)
                  h_added = random.choice(list_to_add)      
                  heuristic_strings[index] = h_added
            neighbors.append(heuristic_strings)

     return neighbors 



# pertubative heuristics for the examination timetabling problem
def Swapactivations(list_activations):
     '''
     swap randomly chosen rows from a timetable
     '''
     row1 = random.choice(list_activations)
     row2 = random.choice(list_activations)
     # swap contents
     indx1 = list_activations.index(row1)
     indx2 = list_activations.index(row2)
     list_activations[indx1] = row2
     list_activations[indx2] = row1
     return list_activations

def allocate_deallocate(list_activations):
     '''
      randomly select an exam from one period and allocate it to another

        'hardswish': nn.Hardswish(),
        'relu6': nn.ReLU6(),
        'elu' : nn.Elu(),
        'hardsigmoid' : nn.Hardsigmoid(),
        'hardtanh'  : nn.Hardtanh(),
        'selu' : nn.Selu()
        'tanh': nn.Tanh
     '''
     static_activations = ['relu','tanh','sigmoid','hardswish','relu6','elu','hardsigmoid','hardtanh','selu','hardshrink','leakyrelu','log_sig','p_relu','r_relu','celu','gelu','softplus','tanshrink']                                       # add more activations

     row1 = random.choice(list_activations)
     inserted = random.choice(static_activations)
     inde = list_activations.index(row1)
     list_activations[inde] = inserted 
     return list_activations  


def shift(list_activations):
           list_activations.insert(0,list_activations.pop()) 
           return list_activations


def insert_shift(list_activations):
      static_activations = ['relu','tanh','sigmoid','hardswish','relu6','elu','hardsigmoid','hardtanh','selu','hardshrink','leakyrelu','log_sig','p_relu','r_relu','celu','gelu','softplus','tanshrink']                                       # add more activations
      row1 = random.choice(list_activations)
      insert = random.choice(static_activations)
      while row1 == insert:
         insert = random.choice(static_activations)
      id  = list_activations.index(row1)

      list_activations[id] = insert
      list_activations.insert(0,list_activations.pop())
      return list_activations


#-- end of heuristics


'''
checks if individual is a string representing a heuristic
'''

def is_heuristic(list_activations,i):
    
    strng_rep = str(i)
    if strng_rep == "h1":
      #print("working with swap activations perturbative heuristic")
      changes = Swapactivations(list_activations)
      return changes
    elif strng_rep == "h2":
      #print("working with allocate/deallocate perturbative heuristic")
      changes = allocate_deallocate(list_activations)
      return changes
    elif strng_rep == "h3" :
          #print("working with shift perturbative heuristic")
          changes = shift(list_activations)
          return changes
    elif strng_rep == "h4" :
          #print("working with insert_shift perturbative heuristic")
          changes = insert_shift(list_activations)
          return changes 

          
# hyper heuristics

def heuristics_train(individual):

    starting_point = ['relu','tanh','relu','tanh','relu','tanh','relu','tanh','relu','tanh','relu','tanh','relu','tanh','relu']
    starting_point_2 = ['relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu','relu']
    for i in individual:
          next_state_activations = is_heuristic(starting_point,i)
    # use the same heuristic string twice to optimise the second larger list of activation functions  
    for i in range(3):
       for i in individual:
         layer_2_activations = is_heuristic(starting_point_2,i)      
    return next_state_activations, layer_2_activations


## fitness function
def fitness(list_of_low_lvl_heuristics):
    ## use low_level_heuristic to construct a timetable and evaluate it, return the objective value as the fitness of the individual.
    list_activations = heuristics_train(list_of_low_lvl_heuristics)
    activationlist_one = list_activations[0]
    act_list_two = list_activations[1]
    epochs = 35
    learning_rate = 0.01  
    #learning_rate = 0.1 original learning rate
    history = fit(epochs,learning_rate,activationlist_one, act_list_two)
    print("--------------------------------- End training with heuristic combination ---------------------------------------")     
    acc_arr = history[0] 
    highest_accuracy = max(acc_arr)
    loss_arr = history[1]
    low_loss = min(loss_arr)
    sm = sum(loss_arr)
    av = sm/len(loss_arr)
    fitness =  low_loss 
    return fitness, highest_accuracy, acc_arr, loss_arr, activationlist_one, act_list_two, av





def tabu_search(
        max_iter: int =20,
        max_size: int = 7
):


    """tabu search algorithm
    given a state, evaluate all its neighbor and
    select the best one while the maximum
    iteration is not reached
    :max_iterations: number of iteration to reach before stopping
    :max_size: tabu list size
    :return: the best solution found which in the form of a heuristic combination and the timetable generated by the combination
    """
    # generating initial state
    s0 = get_initial_state()

    # initializing
    s_best = copy.deepcopy(s0)
    best_candidate = copy.deepcopy(s0)

    # initializing tabu list
    tabu_list = [copy.deepcopy(s0)]
    list_of_times = []
    average_loss_list = []
    end = 0
    for _ in range(max_iter):

        begin = time.time()
        ### time it here 
        s_neighborhood = get_neighbors(best_candidate)
        best_candidate = s_neighborhood[0]

        # for all neighbors
        for s_candidate in s_neighborhood:

            # if we have not already analysed this solution
            if s_candidate in tabu_list:
                continue

            # select the best candidate of all neighbors
            fit_loss_pair_for_s_candidate =  fitness(s_candidate)
            fit_s_candidate = fit_loss_pair_for_s_candidate[0]
            loss_s_candidate = fit_loss_pair_for_s_candidate[1]
            
            fit_loss_pair_for_best_candidate =  fitness(best_candidate)

            fit_best_candidate = fit_loss_pair_for_best_candidate[0]
            
            best_fitness = fit_best_candidate
            best_accuracy = fit_loss_pair_for_best_candidate[1]
            best_acc_arr = fit_loss_pair_for_best_candidate[2]
            best_loss_arr = fit_loss_pair_for_best_candidate[3]

            best_list_activations = fit_loss_pair_for_best_candidate[4]
            best_list_activations_list2 = fit_loss_pair_for_best_candidate[5]
            
            best_average = fit_loss_pair_for_best_candidate[6]
            '''
            average_loss_list.append(best_average)
            '''
            if fit_s_candidate < fit_best_candidate:
                best_candidate = copy.deepcopy(s_candidate)
                fit_best_candidate = fit_loss_pair_for_s_candidate[0]
                best_fitness = fit_best_candidate
                best_accuracy = fit_loss_pair_for_s_candidate[1]
                best_acc_arr = fit_loss_pair_for_s_candidate[2] 
                best_loss_arr = fit_loss_pair_for_s_candidate[3]
                best_list_activations = fit_loss_pair_for_s_candidate[4]
                best_list_activations_list2 = fit_loss_pair_for_s_candidate[5]
                
                best_average = fit_loss_pair_for_s_candidate[6]
                '''
                del average_loss_list[-1]
                average_loss_list.append(best_average)
                '''
        # if the best candidate is better than the best solution
        # it becomes the best solution
        fit_loss_pair_for_s_best = fitness(s_best)
        fit_s_best = fit_loss_pair_for_s_best[0]
        best_accuracy = fit_loss_pair_for_s_best[1]
        best_fitness = fit_s_best
        best_acc_arr = fit_loss_pair_for_s_best[2]
        best_loss_arr = fit_loss_pair_for_s_best[3]
        best_list_activations = fit_loss_pair_for_s_best[4]
        best_list_activations_list2  = fit_loss_pair_for_s_best[5]
        
        best_average =fit_loss_pair_for_s_best[6]
         
        #del average_loss_list[-1]
        average_loss_list.append(best_average)
        
        if fit_best_candidate < fit_s_best:
            s_best = copy.deepcopy(best_candidate)
            fit_best_candidate = fit_loss_pair_for_best_candidate[0]
            best_fitness = fit_best_candidate
            best_accuracy = fit_loss_pair_for_best_candidate[1]
            best_acc_arr = fit_loss_pair_for_best_candidate[2] 
            best_loss_arr = fit_loss_pair_for_best_candidate[3]
            best_list_activations = fit_loss_pair_for_best_candidate[4]
            best_list_activations_list2 = fit_loss_pair_for_best_candidate[5]
            best_average = fit_loss_pair_for_best_candidate[6]
            
            del average_loss_list[-1]
            average_loss_list.append(best_average)
            
        # adding the best candidate to analysed solutions
        tabu_list.append(copy.deepcopy(best_candidate))

        # adjusting tabu list size
        if len(tabu_list) > max_size:
            tabu_list.pop()

        time.sleep(1)
        end = time.time() +  end
        list_of_times.append(end)    
 
    return s_best,best_fitness, best_accuracy, best_acc_arr, best_loss_arr, best_list_activations, best_list_activations_list2, best_average, list_of_times,average_loss_list
```
